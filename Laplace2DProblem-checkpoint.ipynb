```python
def showTurtle(i,j):    
    tur.penup()                # don't draw when turtle moves
    tur.goto(i,j)              # move the turtle to the location (i,j)
    tur.showturtle()           # make the turtle visible
    tur.pendown()              # draw when turtle moves
```

```python
def close(i,j):
    turtle.Screen().bye()                     # close the drawing window
    turtle.TurtleScreen._RUNNING = True       # prevent turtle window initialization errors
```

```python
def startTurtle(grid):    
    global boundaryConditions      # Global variables can be used in all the notebook
    global tur  
    boundaryConditions = grid      # set the matrix of boundary conditions 
    tur= turtle.Turtle()           # creates the turtle object
    tur.shape('circle')            # set the cursor shape as a circle
    tur.pensize(10)                # set the cursor size 
    tur.shapesize(.5,.5,.5)        # set the cursor shape size 
    tur.pencolor('purple')         # set the cursor corlor
    tur.hideturtle()               # hide the cursor
    ws = turtle.Screen()           # initialize the screen for drawing
    ws.setup(300, 300)             # set the screen size in pixels
    ws.screensize(300, 300)        # set the available area for drawing in pixels
    ws.bgcolor('grey')             # set the background color for the window
    tur.speed('fastest')           # set the speed of the cursor
    ws.onclick(showTurtle,1)       # event that calls the function 'showTurtle' when the left button of user mouse is clicked
    tur.ondrag(drag)               # event that calls the function 'drag' when user drags the cursor
    tur.onrelease(message)         # event that calls the 'message' function when the user releases the mouse to start drawing a new line
    ws.onclick(close,3)            # event that calls the 'close' function when the right mouse button is clicked
    ws.mainloop()                  # put everything in a loop
   
```

#### Convolution

It can be shown that the solutions of Laplace's equation satisfy the equality:

$V(x_{i},y_{j})=\frac{1}{4}\left ( V(x_{i+1},y_{j})+ V(x_{i-1},y_{j}) + V(x_{i},y_{j+1})+ V(x_{i},y_{j-1}) \right )$

That is, we don't know the actual solution, but we know that every solution has that property, so we can find each value of the potential in terms of the values at neighboring points. On a 2D grid, the last equation says that we should take the average of the nearest neighbors (above, below, to the left and to the right, excluding diagonal neighbors) to obtain the value of the potential at some point.
The relaxation method is applied to the boundary condition matrix, where we update the matrix by replacing each element with the average of its nearest neighbors. This process is repeated using the updated matrix. After a given number of iterations, the matrix must converge, this is known when the change from one matrix to another is negligible, by the uniqueness of potential theorem, this is the solution to the Laplace equation in that interior region with the specified boundary conditions.

##### A quick introduction can be found in [Griffiths, chapter 3](https://archive.org/details/introduction-to-electrodynamics-d-j-griffiths-4ed).

To take the desired average, we can make a convolution of the matrix of boundary conditions with the next kernel matrix: 

```python
gridUpdated = np.zeros((300,300))
kernel = generate_binary_structure(2,1).astype(float)/4.
kernel[1,1] = 0
kernel               # By the definition of matrix covolution, this will give the desired result
```

The following is the iteration process, where the stopping condition is given by the maximum absolute error between the elements of each array, we accept a solution when this error is less than $10^{-5}$, (it takes a while to converge ).

```python
error=1   # just to enter in the loop
while(error>10**(-5)):
    gridUpdated = convolve(grid, kernel, mode = 'constant',  cval = 0.0)
    gridUpdated[boundaryConditions != 0] = boundaryConditions[boundaryConditions != 0]   # This keep the boundary potential fix on each iteration  
    error = np.max(np.abs(gridUpdated-grid))
    grid = gridUpdated
print('error: ',error)
```
